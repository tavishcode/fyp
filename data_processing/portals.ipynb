```python
import pandas as pd
from tqdm.auto import tqdm
import requests
from bs4 import BeautifulSoup
import re
```

```python
def find_portal(article):
    try:
        res = requests.get("https://en.wikipedia.org/wiki/" + article)
    except:
        return "None"
    soup = BeautifulSoup(res.text, 'html.parser')
    links=soup.find_all(href=re.compile("Portal"))
    portals = ""
    for l in links:
        try:
            portals += l['title'].split('Portal:')[1] + ";"
        except:
            pass
    if portals:
        portals = portals[:-1]
    return portals
```

```python
tqdm.pandas()
df = pd.read_csv('rand_sample.csv', index_col=0)
```

```python
df.head()
```

```python
df.portals= df['link'].progress_apply(find_portal)
```

```python
df.to_csv('portals.csv')
```
