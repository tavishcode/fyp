```python
import requests
from bs4 import BeautifulSoup
import re

def find_portal(article):
    try:
        res = requests.get("https://en.wikipedia.org/wiki/" + article)
    except:
        return "None"
    soup = BeautifulSoup(res.text, 'html.parser')
    links=soup.find_all(href=re.compile("Portal"))
    portals = ""
    for l in links:
        try:
            portals += l['title'].split('Portal:')[1] + ";"
        except:
            pass
    if portals:
        portals = portals[:-1]
    return portals
```

```python
df= df[df['Page'].str.contains('.wikipedia.org')] 
df['lang']= df['Page'].apply(lambda s: s.split(".wikipedia.org")[0][-2:])
```
